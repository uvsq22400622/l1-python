### Filtre vert

Commençons par un filtre qu'on a déjà faité&nbsp;: le filtre vert. Pour ça, copiez la fonction <code class="language-python"><span class="ͼv">filtre_vert</span></code> du TD sur les images. Modifiez-le pour qu'il s'applique sur la variable globale `matrice_pixels`, et finalement, faites en sorte que le callback de l'entrée _Vert_ de votre menu appelle cette fonction.

On pourrait directement associer cette fonction en tant que callback, mais c'est une mauvaise pratique. en programmation, on s'efforce toujours de séparer les responsabilités&nbsp;: d'un côté la gestion de l'interface graphique, de l'autre les traitements d'image. D'ailleurs, un peu plus tard, on sera obligé de faire comme ça...

Mais il manque une dernière étape&nbsp;: une fois ces modifications appliquées, il faut rafraichir la fenêtre, comme quand on ouvre le fichier. Souvenez-vous, l'image est convertie et utilisée dans le canvas. Il faut donc vider ce canvas et le réinitialiser avec la nouvelle image (une `PhotoImage` qui doit être en variable globale elle aussi).

D'ailleurs, l'idéal serait de faire ce  rafraichissement dans une fonction dédiée, parce qu'on en aura sûrement besoin pour d'autres filtres. Il suffira d'appeler cette fonction une fois le filtre appliqué. Et du coup, on peut aussi simplifier l'ouverture de fichier, puisque cette fonction fait la même chose que le rafraichissement. Autant appeler la fonction <code class="language-python"><span class="ͼv">rafraichir</span></code> depuis le callback <code class="language-python"><span class="ͼv">callback_ouvrir</span></code> (ça s'appelle _factoriser le code_). (dans lequel on met canvas.config(ou on vide et rerempli le canvas) on transforme la photoimage en canvas)

Et quitte à réorganiser, autant créer le canvas dès le début (au moment où on crée la fenêtre). Pas la peine de définir sa taille à ce moment-là, il suffit de lui appliquer la taille de la `PhotoImage` quand on rafraichit (un simple appel à <code class="language-python">canvas<span class="ͼ19">.</span><span class="ͼz">config</span></code> puis à <code class="language-python">fenetre_principale<span class="ͼ19">.</span><span class="ͼz">pack_propagate</span></code> fera l'affaire). Ça peut sembler inutile de mettre à jour la taille du canvas à chaque modification (après tout, un filtre vert ne change rien à la taille), mais ça sera utile si certains filtres changent justement cette taille (un zomm, par exemple).

````python
# Définition des variables globales
nom_fichier = None
canvas = None
photo = None
matrice_pixels = None

# Gestion de l'affichage
def rafraichir(matrice_pixels):
    global photo, canvas
    
    photo = ImageTk.PhotoImage(Image.fromarray(matrice_pixels))
    
    if canvas is None:
        # Si c'est la première fois qu'on affiche l'image, on crée le canvas
        canvas = tk.Canvas(fenetre_principale, width = (Image.fromarray(matrice_pixels)).size[0], height = (Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        canvas.pack()
        create=False
    else:
        # Si le canvas a déjà été créé, on le met simplement à jour avec
        # la nouvelle image à afficher
        canvas.delete("all")
        canvas.config(width=(Image.fromarray(matrice_pixels)).size[0], height=(Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)

    fenetre_principale.pack_propagate(True)


# Algorithmes de filtre
def filtre_vert():
    global matrice_pixels
    matrice_pixels[:,:,[0,2]] = 0 

# Callbacks
def callback_vert():
    filtre_vert()
    rafraichir(matrice_pixels)

def callback_ouvrir():
    global photo, nom_fichier, matrice_pixels
    nom_fichier = filedialog.askopenfilename(title="Choisir une image", filetypes=[("Images", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    if nom_fichier:
        image_pil = Image.open(nom_fichier)
        matrice_pixels = np.array(image_pil)
        rafraichir(matrice_pixels)

    #photo = ImageTk.PhotoImage(pil.Image.open(nom_fichier))
    
# Création de la fenêtre principale

fenetre_principale = tk.Tk()
fenetre_principale.title("UVSQolor")

menu_bar = tk.Menu(fenetre_principale)
fenetre_principale.config(menu=menu_bar)

menu_fichier = tk.Menu(menu_bar)
menu_fichier.add_command(
    label="Ouvrir",
    command = callback_ouvrir,
)

menu_effets = tk.Menu(menu_bar)
menu_effets.add_command(
    label="Filtre Vert",
    command=callback_vert,
)

menu_bar.add_cascade(
    label="Fichier",
    menu=menu_fichier, 
)

menu_bar.add_cascade(
    label = "Effets",
    menu=menu_effets
)
fenetre_principale.mainloop()
````

### Filtre gris

Bien, avant de se lancer dans des filtres plus compliqués, avec des interfaces graphiques, commençons par voir si on peut répliquer ce principe avec quelque chose de simple&nbsp;: le passage en niveaux de gris. À votre avis, comment peut-on passer une image couleur en niveaux de gris&nbsp;?

<em> la valeur des 3 composantes doit être la même, la moyenne des trois composantes = perte d'info obligatoire 
pour calculer la luminance = moyenne pondérée par trois coeff (multiplication vectorielle) c1v1+c2v2+c3v3 
np.fot(m1, m2) avec première marice = valeur, deuxième = coeff qui donne nouvelle matrice de pixels </em>

Bien, implémentez le filtre gris, qui va modifier directement la variable globale `matrice_pixels`. Et n'oubliez pas de mettre à jour les menus, pour pouvoir tester ce filtre&nbsp;!

````python
# Définition des variables globales
nom_fichier = None
canvas = None
photo = None
matrice_pixels = None

#def ouvrir():
#    global nom_fichier, photo
#    nom_fichier = filedialog.askopenfilename(title="Ouvrir une image")
#    if nom_fichier is not None:
#        # On crée une image Pillow puis on la convertit au format TkInter 
#        img_pil = pil.Image.open(nom_fichier)
#        img = np.array(img_pil)
#        imf_affichee = img.copy() #copie de la matrice
#
#        photo = ImageTk.PhotoImage(img)
        
#    rafraichir(img_affichee)

# Gestion de l'affichage
def rafraichir(matrice_pixels):
    global photo, canvas
    
    photo = ImageTk.PhotoImage(Image.fromarray(matrice_pixels))
    
    if canvas is None:
        # Si c'est la première fois qu'on affiche l'image, on crée le canvas
        canvas = tk.Canvas(fenetre_principale, width = (Image.fromarray(matrice_pixels)).size[0], height = (Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        canvas.pack()
        create=False
    else:
        # Si le canvas a déjà été créé, on le met simplement à jour avec
        # la nouvelle image à afficher
        canvas.delete("all")
        canvas.config(width=(Image.fromarray(matrice_pixels)).size[0], height=(Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)

    fenetre_principale.pack_propagate(True)   
    
# Algorithmes de filtre
def filtre_gris():
    global matrice_pixels
    if matrice_pixels is not None :
        luminance = np.dot(matrice_pixels[:,:,:3].reshape(-1,3),[0.299,0.587,0.114])
        #reshape = pour réorganiser les pixels de l'image(h,l,3) en (h*l,3)
        luminance_reshape = luminance.reshape(matrice_pixels.shape[0],matrice_pixels.shape[1])
        #réorganise pour mettre sous forme hauteur, largeur
        matrice_pixels_gris = np.stack([luminance_reshape]*3, axis=-1)
        #créer la matrice avec les valeurs de l dans les trois canaux / axis=-1 : les canaux soit le dernier axe d'un tableau
        matrice_pixels = np.clip(matrice_pixels_gris, 0, 255).astype(np.uint8)
        
def filtre_vert():
    global matrice_pixels
    matrice_pixels[:,:,[0,2]] = 0 

# Callbacks
def callback_vert():
    filtre_vert()
    rafraichir(matrice_pixels)

def callback_ouvrir():
    global photo, nom_fichier, matrice_pixels
    nom_fichier = filedialog.askopenfilename(title="Choisir une image", filetypes=[("Images", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    if nom_fichier:
        image_pil = Image.open(nom_fichier)
        matrice_pixels = np.array(image_pil)
        rafraichir(matrice_pixels)

def callback_gris():
    filtre_gris()
    rafraichir(matrice_pixels)

# Création de la fenêtre principale
fenetre_principale = tk.Tk()
fenetre_principale.title("UVSQolor")

menu_bar = tk.Menu(fenetre_principale)
fenetre_principale.config(menu=menu_bar)

menu_fichier = tk.Menu(menu_bar)
menu_fichier.add_command(
    label="Ouvrir",
    command = callback_ouvrir,
)

menu_effets = tk.Menu(menu_bar)
menu_effets.add_command(
    label="Filtre Vert",
    command=callback_vert,
)

menu_effets.add_command(
    label="Filtre Gris",
    command=callback_gris,
)

menu_bar.add_cascade(
    label="Fichier",
    menu=menu_fichier, 
)

menu_bar.add_cascade(
    label = "Effets",
    menu=menu_effets
)

# Boucle principale
fenetre_principale.mainloop()
````

    * <code class="language-python"><span class="ͼv">applique_effet</span></code> va valider le changement. Il suffit de recopier l'image affichée dans l'image d'origine. Il faut ensuite fermer la boite de dialogue.
    * <code class="language-python"><span class="ͼv">annule_effet</span></code>&nbsp; va faire l'inverse, recopier l'image d'origine dans l'image affichée (attention, dans ce cas, il faut aussi rafraichir) et fermer la boite de dialogue.



````python
# Définition des variables globales
matrice_pixels_affichee = None
nom_fichier = None
canvas = None
photo = None
matrice_pixels = None


# Gestion de l'affichage
def rafraichir(matrice_pixels):
    global photo, canvas
    
    photo = ImageTk.PhotoImage(Image.fromarray(matrice_pixels))
    
    if canvas is None:
        # Si c'est la première fois qu'on affiche l'image, on crée le canvas
        canvas = tk.Canvas(fenetre_principale, width = (Image.fromarray(matrice_pixels)).size[0], height = (Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        canvas.pack()
        create=False
    else:
        # Si le canvas a déjà été créé, on le met simplement à jour avec
        # la nouvelle image à afficher
        canvas.delete("all")
        canvas.config(width=(Image.fromarray(matrice_pixels)).size[0], height=(Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)

    fenetre_principale.pack_propagate(True)   
    
def ouvrir_dialogue_luminosite():
    global dialogue_effet
    dialogue_effet = tk.Toplevel(fenetre_principale)
    dialogue_effet.title("Luminosité")
    dialogue_effet.geometry("300x150")
    dialogue_effet.grab_set()
    slider = tk.Scale(dialogue_effet, from_=0.1, to=3.0,
                      orient=tk.HORIZONTAL, length=200,
                      resolution=0.1, digits=2,
                      command=correction_gamma)
    #scale = le glisseur
    #appliquer un symétrique sur from et to / initialisé le slider.set(0) pour le faire, linéarisation requise
    slider.set(1.0)
    slider.pack(pady=20)

    frame_boutons = tk.Frame(dialogue_effet)
    frame_boutons.pack(side=tk.BOTTOM, pady=10)

    bouton_appliquer = tk.Button(frame_boutons, text="Appliquer",
                                 command=applique_effet)
    bouton_appliquer.pack(side=tk.LEFT, padx=10)

    bouton_annuler = tk.Button(frame_boutons, text="Annuler",
                               command=annule_effet)
    bouton_annuler.pack(side=tk.LEFT, padx=10)

def applique_effet():
    global matrice_pixels, matrice_pixels_affichee, dialogue_effet
    if matrice_pixels_affichee is not None:
        matrice_pixels = matrice_pixels_affichee.copy()
        #rafraichir(matrice_pixels) pas obligatoire puisque déjà affichée
    if dialogue_effet:
        dialogue_effet.destroy() 

def annule_effet():
    global matrice_pixels, matrice_pixels_affichee, dialogue_effet
    if matrice_pixels is not None:
        matrice_pixels_affichee = matrice_pixels.copy()
        rafraichir(matrice_pixels_affichee)
    if dialogue_effet:
        dialogue_effet.destroy()
#rajouter un callback pour quitter la fenetre luminosite qui annule les changements

# Algorithmes de filtre
def filtre_gris():
    global matrice_pixels
    if matrice_pixels is not None :
        luminance = np.dot(matrice_pixels[:,:,:3].reshape(-1,3),[0.299,0.587,0.114])
        #reshape = pour réorganiser les pixels de l'image(h,l,3) en (h*l,3)
        luminance_reshape = luminance.reshape(matrice_pixels.shape[0],matrice_pixels.shape[1])
        #réorganise pour mettre sous forme hauteur, largeur
        matrice_pixels_gris = np.stack([luminance_reshape]*3, axis=-1)
        #créer la matrice avec les valeurs de l dans les trois canaux / axis=-1 : les canaux soit le dernier axe d'un tableau
        matrice_pixels = np.clip(matrice_pixels_gris, 0, 255).astype(np.uint8)

def filtre_vert():
    global matrice_pixels
    matrice_pixels[:,:,[0,2]] = 0 

def correction_gamma(gamma_str):
    global matrice_pixels, matrice_pixels_affichee
    if matrice_pixels is not None:
        gamma = float(gamma_str)
        max_value = float(np.iinfo(matrice_pixels.dtype).max)
        #float rajoute de la précision
        img_normalisee = matrice_pixels / max_value
        img_ajustee = np.power(img_normalisee, gamma)
        #np.power(matrice, puissance)applique la puissance gamma à la matrice
        matrice_pixels_affichee = (img_ajustee * max_value).astype(matrice_pixels.dtype)
        #.astype() permet de changer le type de donnée en un autre spécifié (np.uint8) dans l'ancienne version du gamma TP4
        rafraichir(matrice_pixels_affichee)

#pour linéariser : gamma = f exposant ln
#ln_arr = np.log(arr)

# Callbacks
def callback_vert():
    filtre_vert()
    rafraichir(matrice_pixels)

def callback_ouvrir():
    global photo, nom_fichier, matrice_pixels
    nom_fichier = filedialog.askopenfilename(title="Choisir une image", filetypes=[("Images", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    if nom_fichier:
        image_pil = Image.open(nom_fichier)
        matrice_pixels = np.array(image_pil)
        rafraichir(matrice_pixels)

def callback_gris():
    filtre_gris()
    rafraichir(matrice_pixels)

def callback_luminosite():
    luminosite()
def callback_correction_gamma():
    correction_gamma()

# Création de la fenêtre principale
fenetre_principale = tk.Tk()
fenetre_principale.title("UVSQolor")

menu_bar = tk.Menu(fenetre_principale)
fenetre_principale.config(menu=menu_bar)

menu_fichier = tk.Menu(menu_bar)
menu_fichier.add_command(
    label="Ouvrir",
    command = callback_ouvrir,
)

menu_effets = tk.Menu(menu_bar)
menu_effets.add_command(
    label="Filtre Vert",
    command=callback_vert,
)

menu_effets.add_command(
    label="Filtre Gris",
    command=callback_gris,
)

menu_bar.add_cascade(
    label="Fichier",
    menu=menu_fichier, 
)

menu_bar.add_cascade(
    label = "Effets",
    menu=menu_effets
)

menu_effets.add_command(
    label="Luminosité", 
    command=ouvrir_dialogue_luminosite
)

# Boucle principale
fenetre_principale.mainloop()
````

Bien, ça marche.


### Contraste

Maintenant, on va implémenter un nouveau filtre, qui va ajuster le contraste. Cette fois, on va utiliser une [fonction sigmoïde](https://fr.wikipedia.org/wiki/Sigmo%C3%AFde_(math%C3%A9matiques)). L'idée n'est pas bien compliquée, une fois qu'on a à peu près compris le principe mathématique. Grace à sa courbe en S, les valeurs moyennes sont étirées (partie centrale de la courbe en S avec une pente plus raide), tandis que les valeurs extrêmes sont compressées (parties aplaties aux extrémités de la courbe en S), ce qui accentue les différences dans les tons moyens tout en préservant les détails dans les hautes lumières et les ombres.

Quand on utilise cette fonction $f(x)=\frac {1}{1+{\rm {e}}^{-x}}$, sur une image, on a&nbsp;:
 * $x$ est la valeur de pixel d'origine, normalisée entre 0 et 1
 * $f(x)$ est la nouvelle valeur de pixel.

Pour adapter cette fonction sigmoïde standard à un ajustement de contraste, nous procédons comme suit&nbsp;:
 * Normalisation<br>Nous normalisons les valeurs de pixels $x$ dans l'intervalle $[0, 1]$.
 * Centrage<br>Nous centrons la fonction autour de 0.5 (pour que les valeurs de pixel médianes ne soient pas affectées par la transformation) en soustrayant $0.5$ à $x$&nbsp;: $(x - 0.5)$.
 * Ajout du contraste : Nous la multiplions par le facteur de contraste $c$&nbsp;: $c(x - 0.5)$ (vous noterez que ce facteur correspond au $\lambda$ de la forme généralisée de la sigmoïde).
* Ajout de la pente : Nous introduisons une pente $k$ pour contrôler la transition, résultant en $k \times c(x - 0.5)$. À l'usage, une pente de 10 semble appropriée, mais vous pouvez essayer d'autres valeurs (voire permettre le contrôle de la pente par l'utilisateur).

La formule finale (sur un $x$ déjà normalisé à $[0,1]$) devient donc&nbsp;:
$$f(x)=\frac {1}{1+{\rm {e}}^{-k \times c(x-0,5)}}$$

Cette fois, on a deux paramètres su lesquels jouer&nbsp;: la pente, et le contraste. On va ajouter un deuxième slider. Par contre, comme notre correction de contraste a besoin des deux paramètres, il faudra faire en sorte que chaque fois que l'un des deux sliders est déplacé, la fonction reçoive les deux valeurs. Voici comment indiquer ça dans le paramètre `command` d'un sliderénbsp;:
```python
    command=lambda x: correction_sigmoide(slider1.get(), slider2.get())
```
Ce code passe comme callback une _expression lambda_, c'est-à-dire une fonction directement créée par l'instruction. Ça revient en gros à créer une fonction
```python
def expression_lambda():
    correction_sigmoide(slider1.get(), slider2.get())
```

et à la passer comme paramètre `command`. Sauf qu'on n'a pas besoin de la créer _puis_ de l'utiliser. On la crée directement au moment de la passer en paramètre `command`. C'est assez pratique.

````python
# Définition des variables globales
matrice_pixels_affichee = None
nom_fichier = None
canvas = None
photo = None
matrice_pixels = None


# Gestion de l'affichage
def rafraichir(matrice_pixels):
    global photo, canvas
    
    photo = ImageTk.PhotoImage(Image.fromarray(matrice_pixels))
    
    if canvas is None:
        # Si c'est la première fois qu'on affiche l'image, on crée le canvas
        canvas = tk.Canvas(fenetre_principale, width = (Image.fromarray(matrice_pixels)).size[0], height = (Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        canvas.pack()
        create=False
    else:
        # Si le canvas a déjà été créé, on le met simplement à jour avec
        # la nouvelle image à afficher
        canvas.delete("all")
        canvas.config(width=(Image.fromarray(matrice_pixels)).size[0], height=(Image.fromarray(matrice_pixels)).size[1])
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)

    fenetre_principale.pack_propagate(True)   
    
def ouvrir_dialogue_luminosite():
    global dialogue_effet
    dialogue_effet = tk.Toplevel(fenetre_principale)
    dialogue_effet.title("Luminosité")
    dialogue_effet.geometry("300x150")
    dialogue_effet.grab_set()
    slider = tk.Scale(dialogue_effet, from_=0.1, to=3.0,
                      orient=tk.HORIZONTAL, length=200,
                      resolution=0.1, digits=2,
                      command=correction_gamma)
    slider.set(1.0)
    slider.pack(pady=20)

    frame_boutons = tk.Frame(dialogue_effet)
    frame_boutons.pack(side=tk.BOTTOM, pady=10)

    bouton_appliquer = tk.Button(frame_boutons, text="Appliquer",
                                 command=applique_effet)
    bouton_appliquer.pack(side=tk.LEFT, padx=10)

    bouton_annuler = tk.Button(frame_boutons, text="Annuler",
                               command=annule_effet)
    bouton_annuler.pack(side=tk.LEFT, padx=10)

def correction_gamma(gamma_str):
    global matrice_pixels, matrice_pixels_affichee
    if matrice_pixels is not None:
        gamma = float(gamma_str)
        max_value = float(np.iinfo(matrice_pixels.dtype).max)
        img_normalisee = matrice_pixels / max_value
        img_ajustee = np.power(img_normalisee, gamma)
        matrice_pixels_affichee = (img_ajustee * max_value).astype(matrice_pixels.dtype)
        rafraichir(matrice_pixels_affichee)

def ouvrir_dialogue_contraste():
    global dialogue_effet
    dialogue_effet = tk.Toplevel(fenetre_principale)
    dialogue_effet.title("Contraste")
    dialogue_effet.geometry("300x150")
    dialogue_effet.grab_set()

    slider1 = tk.Scale(dialogue_effet, from_=1.2, to=20.0, 
                        orient=tk.HORIZONTAL, length=200,
                        resolution=0.1, digits=2,
                        command=lambda x: correction_sigmoide(slider1.get(), slider2.get()))
    slider1.set(10.0)
    slider1.pack(pady=10)

    slider2 = tk.Scale(dialogue_effet, from_=1.2, to=20.0,
                        orient=tk.HORIZONTAL, length=200,
                        resolution=0.1, digits=2,
                        command=lambda x: correction_sigmoide(slider1.get(), slider2.get()))
    slider2.set(10.0)
    slider2.pack(pady=10)

    frame_boutons = tk.Frame(dialogue_effet)
    frame_boutons.pack(side=tk.BOTTOM, pady=10)

    bouton_appliquer = tk.Button(frame_boutons, text="Appliquer", command=applique_effet)
    bouton_appliquer.pack(side=tk.LEFT, padx=10)

    bouton_annuler = tk.Button(frame_boutons, text="Annuler", command=annule_effet)
    bouton_annuler.pack(side=tk.LEFT, padx=10)

def correction_sigmoide(contraste, pente):
    global matrice_pixels, matrice_pixels_affichee
    if matrice_pixels is not None:
        #normalisation:
        x = np.clip(matrice_pixels, 0, 255) / 255.0
        #transformation sigmoide centrée autour de 0.5
        x = x - 0.5 #centrage
        contraste = float(contraste)
        pente = float(pente)
        #calcul de la fonction
        exposant = (-pente) * contraste * x
        img_ajustee = 1 / (1 + np.exp(exposant))
        #conversion de l'image ajustée entre 0 et 255
        matrice_pixels_affichee = (img_ajustee * 255).astype(np.uint8)
        rafraichir(matrice_pixels_affichee)

def applique_effet():
    global matrice_pixels, matrice_pixels_affichee, dialogue_effet
    if matrice_pixels_affichee is not None:
        matrice_pixels = matrice_pixels_affichee.copy()
        rafraichir(matrice_pixels)
    if dialogue_effet:
        dialogue_effet.destroy() 

def annule_effet():
    global matrice_pixels, matrice_pixels_affichee, dialogue_effet
    if matrice_pixels is not None:
        matrice_pixels_affichee = matrice_pixels.copy()
        rafraichir(matrice_pixels_affichee)
    if dialogue_effet:
        dialogue_effet.destroy()

def filtre_gris():
    global matrice_pixels
    if matrice_pixels is not None :
        luminance = np.dot(matrice_pixels[:,:,:3].reshape(-1,3),[0.299,0.587,0.114])
        #reshape = pour réorganiser les pixels de l'image(h,l,3) en (h*l,3)
        luminance_reshape = luminance.reshape(matrice_pixels.shape[0],matrice_pixels.shape[1])
        #réorganise pour mettre sous forme hauteur, largeur
        matrice_pixels_gris = np.stack([luminance_reshape]*3, axis=-1)
        #créer la matrice avec les valeurs de l dans les trois canaux / axis=-1 : les canaux soit le dernier axe d'un tableau
        matrice_pixels = np.clip(matrice_pixels_gris, 0, 255).astype(np.uint8)

def filtre_vert():
    global matrice_pixels
    matrice_pixels[:,:,[0,2]] = 0 

# Callbacks
def callback_vert():
    filtre_vert()
    rafraichir(matrice_pixels)

def callback_ouvrir():
    global photo, nom_fichier, matrice_pixels
    nom_fichier = filedialog.askopenfilename(title="Choisir une image", filetypes=[("Images", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    if nom_fichier:
        image_pil = Image.open(nom_fichier)
        matrice_pixels = np.array(image_pil)
        rafraichir(matrice_pixels)

def callback_gris():
    filtre_gris()
    rafraichir(matrice_pixels)

def callback_luminosite():
    luminosite()
def callback_correction_gamma():
    correction_gamma()

# Création de la fenêtre principale
fenetre_principale = tk.Tk()
fenetre_principale.title("UVSQolor")

menu_bar = tk.Menu(fenetre_principale)
fenetre_principale.config(menu=menu_bar)

menu_fichier = tk.Menu(menu_bar)
menu_fichier.add_command(
    label="Ouvrir",
    command = callback_ouvrir,
)

menu_effets = tk.Menu(menu_bar)
menu_effets.add_command(
    label="Filtre Vert",
    command=callback_vert,
)

menu_effets.add_command(
    label="Filtre Gris",
    command=callback_gris,
)

menu_bar.add_cascade(
    label="Fichier",
    menu=menu_fichier, 
)

menu_bar.add_cascade(
    label = "Effets",
    menu=menu_effets
)

menu_effets.add_command(
    label="Luminosité", 
    command=ouvrir_dialogue_luminosite
)

menu_effets.add_command(
    label="Contraste",
    command=ouvrir_dialogue_contraste
)

# Boucle principale
fenetre_principale.mainloop()
````

### Flou

On va maintenant implémenter un _flou_. À votre avis, comment faire ? Imaginez un pixel noir entouré de 8 pixels blancs. Quelle couleur voudriez-vous donner à ce pixel&nbsp;? Quelle opération simple permet de l'obtenir&nbsp;?

<em>moyenne des deux canaux de lumière, pour que le pixel noir paraisse gris</em>

Ça tombe bien, il existe une opération qui fait exactement ça&nbsp;: la _convolution_.

Une opération de convolution consiste à remplacer la valeur d'un emplacement dans une matrice par une moyenne pondérée des valeurs des emplacements voisins. Cette pondération est définie par une autre matrice, appelée matrice de convolution ou, plus simplement, noyau. Chaque valeur du noyau agit comme un coefficient qui détermine l'importance relative des voisins. Le centre du noyau correspond à l'emplacement traité dans la matrice initiale, tandis que les autres coefficients s'appliquent aux emplacements voisins.

Ici, la valeur $\color{red}c_{u}$ sera remplacée par $\color{red}c_{u}\cdot u\color{black} ~+~ \color{blue}c_{v_1}\cdot v_1\color{black} ~+~ \color{blue}c_{v_2}\cdot v_2 \color{black}~+~ \color{blue}c_{v_3}\cdot v_3 \color{black}~+~ \color{blue}c_{v_4}\cdot v_4 \color{black}~+~ \color{blue}c_{v_5}\cdot v_5 \color{black}~+~ \color{blue}c_{v_6}\cdot v_6 \color{black}~+~ \color{blue}c_{v_7}\cdot v_7 \color{black}~+~ \color{blue}c_{v_8}\cdot v_8$.

Donc si on applique ce noyau aux 3 canaux de l'ensemble des pixels, on devrait flouter (oui, c'est comme ça que ça se dit) l'intégralité de l'image. On pourrait implémenter ce calcul à la main, mais c'est un peu pénible car il faut gérer le cas particulier des bords (avec les coins, ça fait 8 cas particuliers...). Ça peut se faire automatiquement avec la fonction <code class="language-python">np.<span class="ͼv">pad</span></code> qui ajoute des données aux bords de la matrice, justement pour faire face à ce genre de situation. Mais il existe une fonction déjà implémentée. Elle ne se trouve pas dans NumPy, mais dans SciPy, une autre bibliothèque très connue de calcul scientifique. Et elle a été implémentée pour être compatible avec les array NumPy.

Pour les curieux, voici tout de même à quoi ressemblerait une implémentation manuelle de la convolution&nbsp;:

```python
def convolve(image, kernel):
    """
    Applique une convolution 2D sur une image avec un noyau donné.
    
    :param image: Image 2D (grayscale) ou un canal d'image (array NumPy).
    :param kernel: Noyau de convolution (array NumPy 2D).
    :return: Image convoluée (array NumPy 2D).
    """
    # Dimensions de l'image et du noyau
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape

    # Calcul des marges pour le padding
    pad_h = kernel_height // 2
    pad_w = kernel_width // 2

    # Ajouter un padding autour de l'image
    padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')

    # Créer une image de sortie vide
    output = np.zeros_like(image)

    # Appliquer la convolution
    for i in range(image_height):
        for j in range(image_width):
            # Extraire la région locale autour du pixel
            region = padded_image[i:i + kernel_height, j:j + kernel_width]
            # Calculer le produit élément par élément et la somme
            output[i, j] = np.sum(region * kernel)
    
    return output
```

Ce n'est finalement pas très compliqué. Mais puisqu'on a déjà une [fonction toute faite](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.convolve2d.html) (et implémentée en C, donc beaucoup plus rapide), autant l'utiliser.

Commencez par importer le module nécessaire&nbsp;:
```python
from scipy.signal import convolve2d
```
Ensuite, il suffit de passer les bonnes informations à la fonction <code class="language-python"><span class="ͼv">convolve2D</span></code>&nbsp;:
* La matrice de départ
* le kernel
* quelques options qu'on détaille plus bas.

Pour la matrice de départ, c'est un peu subtil. On veut faire une convolution _canal par canal_. Si on utilisait des listes Python, ce serait compliqué. Mais avec des NumPy array, c'est très facile, grace à ses capacités d'indéxation avancée. Prenez l'expression&nbsp;:
```python
image[:,:,i]
```
Avec un array NumPy, c'est une tranche 2D de l'image pour l'indice `i`. Ça veut dire que cette expression permet de récupérer l'intégralité de l'image, uniquement sur le canal `i`, comme si les autres canaux avaient été supprimés.

Donc il suffit de boucler sur le nombre de canaux, de calculer à chaque fois la convolution. Mais une fois qu'on a appliqué la convolution au tableau contenant uniquement le canal `i`, comment réassembler le tout&nbsp;? Là aussi, ce serait compliqué avec des listes Python, mais très simple avec des array NumPy. L'instruction&nbsp;:
```python
image[:,:,i] = ...
```

### Undo
Il y a une fonction absolument essentielle dans n'importe quel logiciel&nbsp;: la fonction _Annuler_&nbsp;!!!

Annuler une action, ce n'est pas très compliqué, il suffit de stocker l'image avant application du filtre, et la restituer en cas d'annulation. Mais dans la plupart des logiciels, on peut annuler _et refaire_ plusieurs fois. Comment peut-on s'y prendre&nbsp;?

La première chose à comprendre, c'est qu'on devra stocker en mémoire chaque version de la matrice de pixels, et charger la version correspondant à l'action _Annuler_/_Refaire_.<br>

Ensuite, on pourrait penser à l'utilisation d'une pile. Le problème est que, si ça permet de revenir en arrière en dépilant les anciennes versions, ça ne permet pas de revenir en avant, puisque le dépilement provoque une suppression des données. Pour résoudre ce problème, on va utiliser une liste. Et comme  l'accès au dernier élément est en $O(1)$ que ce soit pour la liste ou le deque, on aura tout de même un comportement efficace en terme de rapidité avec une liste.

une fois définie cette structure, il faut&nbsp;:
 * mettre en place des variables globales `historique` et `indice_historique`. Cette dernière variable désigne la version courante. Elle permet de revenir en arrière à coup d'_Annulation_ sans pour autant supprimer les transformations que des _Rétablir_ permettraient de recharger.
 * Implémenter des fonctions `annuler` et `retablir` qui mettent à jour l'historique et l'indice, et rafraichissent l'image
 * Gérer la mise à jour de l'historique&nbsp;:
     * À chaque annulation on rétablissement, on va se déplacer dans l'historique, en modifiant l'indice et en rafraichissant l'affichage
     * En cas d'application d'un filtre, il faut ajouter la version qui a été modifiée, mais aussi supprimer de la liste les éventuels historiques qui avaient été annulés, et qui sont écrasés par cette nouvelle modification&nbsp;:

Avec cette organisation, on n'a plus besoin d'avoir deux matrices. La matrice affichée est la variable globale seule, et l'ensemble des images d'origine sont dans la liste. Quand on appliquera un filtre avec prévisualisation (comme la correction gamma), il faudra prendre comme image d'origine celle dans `historique[indice_historique]`.
